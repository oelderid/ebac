#### **Tratando valores ausentes com zero** <a name="Tratando-valores-ausentes-com-zero"></a>

Nessa etapa serão tratados os valores ausentes das variáveis binárias e também das variáveis condomínio, iptu, andar, quartos e vagas. Os valores ausentes dessas variáveis podem ser substituídos por zero, pois ele indica o valor oposto no caso das variáveis e ausência, no caso das outras variáveis mencionadas.



```python
# obtendo o nome das colunas do DataFrame para as quais esse tratamento será aplicado
aplicaveis = pe.get_columns_except([
    'bairro',
    'tipo',
    'valor',
    'metro_trem',
    'metro_quadrado'
])
```

```python
# preenchendo os valores ausentes com 0 para as variáveis binárias e também das variáveis andar, quartos e vagas
df[aplicaveis] = df[aplicaveis].fillna(0)
```

#### **Tratando valores ausentes na variável metro quadrado** <a name="Tratando-valores-ausentes-na-variável-metro-quadrado"></a>

Conforme verificado anteriormente, existem alguns registros na base de dados com valor ausente para a variável "metro_quadrado". Esses registros serão listados abaixo a fim de verificar a melhor forma de tratamento.



```python
df[df['metro_quadrado'].isna()][['bairro', 'tipo', 'valor', 'metro_trem', 'metro_quadrado']]
```

Podemos verificar que apenas 5 registros não possuem o metro quadrado, todos eles do tipo terreno. Pela baixa quantidade de registros e por se tratar de uma variável importante para definição do preço do aluguel, ainda mais se tratando de um terreno, cuja área bruta locável é importante para o inquilino, podemos considerar o descarte desses registros.



```python
# removendo os registros com dados ausentes na varíavel 'metro_quadrado'
registros_remover = df[df['metro_quadrado'].isna()].index
df.drop(index=registros_remover, inplace=True)
```

#### **Alterando os tipos das variáveis** <a name="Alterando-os-tipos-das-variáveis"></a>

Nessa etapa o tipo das variáveis binárias será alterado de float para int, uma vez que o valor desse tipo de variável só pode 1 para representar sim e 0 para não. Também devemos observar os valores das variáveis condomínio e iptu. Esses valores podem não ser informados com casas decimais, dessa forma poderiam também ter o tipo alterado para int. Abaixo será realizado um teste para verificar se essas duas variáveis possuem apenas valores inteiros ou não.

```python
# Verificando se a variável condominio possui de fato apenas valores inteiros
np.array_equal(df.condominio, df.condominio.astype(int))
```

```python
# Verificando se a variável iptu possui de fato apenas valores inteiros
np.array_equal(df.iptu, df.iptu.astype(int))
```

<div style="border: 1px solid #88AB8E; background-color: #ECF4E3; padding: 10px; border-radius: 5px;">
    Como podemos observar no teste anterior, as variáveis condominio e iptu só possuem valores inteiros. O tipo dessas variáveis também pode ser alterado de float para int.
</div>

```python
aplicaveis = pe.get_columns_except([
    'bairro',
    'tipo',
    'valor',
    'metro_trem'
])
```

```python
df[aplicaveis] = df[aplicaveis].astype(int)
```

```python
pe.info()
```

<div style="border: 1px solid #88AB8E; background-color: #ECF4E3; padding: 10px; border-radius: 5px;">
    Podemos verificar acima que agora não existem dados ausentes e também todos os tipos estão definidos corretamente.


</div>

#### **Verificando variáveis do tipo texto** <a name="Verificando-variáveis-do-tipo-texto"></a>

Nessa etapa serão analisados os valores das variáveis bairro e tipo, em busca de valores duplicados ou de erros ortográficos. Abaixo é exibido uma tabela com o nome de todos os bairros ordenados alfabeticamente.

```python
pe.list_column('bairro', 7)
```

<div style="border: 1px solid #88AB8E; background-color: #ECF4E3; padding: 10px; border-radius: 5px;">
    Na tabela acima podemos ver que existem alguns bairros se repetem porque existem caracteres diferentes entre os nomes. A listagem abaixo apresenta os casos que podemos observar:
<ul>
    <li>Barra <b>D</b>e Guaratiba e Barra de Guaratiba</li>
    <li>Braz de Pina e Br<b>ás</b> de Pina</li>
    <li>Engenho <b>D</b>e Dentro e Engenho de Dentro</li>
    <li>Estacio e Est<b>á</b>cio</li>
    <li><b>Freguesia</b> como Freguesia (Jacarepaguá) e Freguesia- Jacarepaguá</li>
    <li>Recreio <b>D</b>os Bandeirantes e Recreio dos Bandeirantes</li>
</ul>

Também podemos ver que existe um bairro denominado <b>Rio de Janeiro - RJ</b>.<br>
Claramente esse não é um nome de um bairro e os registros com esse bairro não estão corretos.
</div>

Abaixo é feita a correção dos nomes dos bairros cadastrados incorretamente na base, padronizando o nome desses bairros.

```python
df['bairro'] = df['bairro'].replace({
    'Barra De Guaratiba': 'Barra de Guaratiba',
    'Braz de Pina': 'Brás de Pina',
    'Engenho De Dentro': 'Engenho de Dentro',
    'Estacio': 'Estácio',
    'Freguesia (Jacarepaguá)': 'Freguesia',
    'Freguesia- Jacarepaguá': 'Freguesia',
    'Recreio Dos Bandeirantes': 'Recreio dos Bandeirantes',
})
```

Abaixo verificamos quantos registros existem na base com o bairro denominado <b>Rio de Janeiro - RJ</b>

```python
df[df['bairro'] == 'Rio de Janeiro - RJ'].shape[0]
```

Considerando que bairro é uma variável importante para a análise e que existem apenas 28 registros com essa classificação, podemos removê-los da base de dados, o que será feito logo abaixo:

```python
df.drop(df[df['bairro'] == 'Rio de Janeiro - RJ'].index, inplace=True)
```
