#### **Tratando valores ausentes com zero** <a name="Tratando-valores-ausentes-com-zero"></a>

Nessa etapa sero tratados os valores ausentes das variveis binrias e tambm das variveis condomnio, iptu, andar, quartos e vagas. Os valores ausentes dessas variveis podem ser substitudos por zero, pois ele indica o valor oposto no caso das variveis e ausncia, no caso das outras variveis mencionadas.



```python
# obtendo o nome das colunas do DataFrame para as quais esse tratamento ser aplicado
aplicaveis = pe.get_columns_except([
    'bairro',
    'tipo',
    'valor',
    'metro_trem',
    'metro_quadrado'
])
```

```python
# preenchendo os valores ausentes com 0 para as variveis binrias e tambm das variveis andar, quartos e vagas
df[aplicaveis] = df[aplicaveis].fillna(0)
```

#### **Tratando valores ausentes na varivel metro quadrado** <a name="Tratando-valores-ausentes-na-varivel-metro-quadrado"></a>

Conforme verificado anteriormente, existem alguns registros na base de dados com valor ausente para a varivel "metro_quadrado". Esses registros sero listados abaixo a fim de verificar a melhor forma de tratamento.



```python
df[df['metro_quadrado'].isna()][['bairro', 'tipo', 'valor', 'metro_trem', 'metro_quadrado']]
```

Podemos verificar que apenas 5 registros no possuem o metro quadrado, todos eles do tipo terreno. Pela baixa quantidade de registros e por se tratar de uma varivel importante para definio do preo do aluguel, ainda mais se tratando de um terreno, cuja rea bruta locvel  importante para o inquilino, podemos considerar o descarte desses registros.



```python
# removendo os registros com dados ausentes na varavel 'metro_quadrado'
registros_remover = df[df['metro_quadrado'].isna()].index
df.drop(index=registros_remover, inplace=True)
```

#### **Alterando os tipos das variveis** <a name="Alterando-os-tipos-das-variveis"></a>

Nessa etapa o tipo das variveis binrias ser alterado de float para int, uma vez que o valor desse tipo de varivel s pode 1 para representar sim e 0 para no. Tambm devemos observar os valores das variveis condomnio e iptu. Esses valores podem no ser informados com casas decimais, dessa forma poderiam tambm ter o tipo alterado para int. Abaixo ser realizado um teste para verificar se essas duas variveis possuem apenas valores inteiros ou no.

```python
# Verificando se a varivel condominio possui de fato apenas valores inteiros
np.array_equal(df.condominio, df.condominio.astype(int))
```

```python
# Verificando se a varivel iptu possui de fato apenas valores inteiros
np.array_equal(df.iptu, df.iptu.astype(int))
```

<div style="border: 1px solid #88AB8E; background-color: #ECF4E3; padding: 10px; border-radius: 5px;">
    Como podemos observar no teste anterior, as variveis condominio e iptu s possuem valores inteiros. O tipo dessas variveis tambm pode ser alterado de float para int.
</div>

```python
aplicaveis = pe.get_columns_except([
    'bairro',
    'tipo',
    'valor',
    'metro_trem'
])
```

```python
df[aplicaveis] = df[aplicaveis].astype(int)
```

```python
pe.info()
```

<div style="border: 1px solid #88AB8E; background-color: #ECF4E3; padding: 10px; border-radius: 5px;">
    Podemos verificar acima que agora no existem dados ausentes e tambm todos os tipos esto definidos corretamente.


</div>

#### **Verificando variveis do tipo texto** <a name="Verificando-variveis-do-tipo-texto"></a>

Nessa etapa sero analisados os valores das variveis bairro e tipo, em busca de valores duplicados ou de erros ortogrficos. Abaixo  exibido uma tabela com o nome de todos os bairros ordenados alfabeticamente.

```python
pe.list_column('bairro', 7)
```

<div style="border: 1px solid #88AB8E; background-color: #ECF4E3; padding: 10px; border-radius: 5px;">
    Na tabela acima podemos ver que existem alguns bairros se repetem porque existem caracteres diferentes entre os nomes. A listagem abaixo apresenta os casos que podemos observar:
<ul>
    <li>Barra <b>D</b>e Guaratiba e Barra de Guaratiba</li>
    <li>Braz de Pina e Br<b>s</b> de Pina</li>
    <li>Engenho <b>D</b>e Dentro e Engenho de Dentro</li>
    <li>Estacio e Est<b></b>cio</li>
    <li><b>Freguesia</b> como Freguesia (Jacarepagu) e Freguesia- Jacarepagu</li>
    <li>Recreio <b>D</b>os Bandeirantes e Recreio dos Bandeirantes</li>
</ul>

Tambm podemos ver que existe um bairro denominado <b>Rio de Janeiro - RJ</b>.<br>
Claramente esse no  um nome de um bairro e os registros com esse bairro no esto corretos.
</div>

Abaixo  feita a correo dos nomes dos bairros cadastrados incorretamente na base, padronizando o nome desses bairros.

```python
df['bairro'] = df['bairro'].replace({
    'Barra De Guaratiba': 'Barra de Guaratiba',
    'Braz de Pina': 'Brs de Pina',
    'Engenho De Dentro': 'Engenho de Dentro',
    'Estacio': 'Estcio',
    'Freguesia (Jacarepagu)': 'Freguesia',
    'Freguesia- Jacarepagu': 'Freguesia',
    'Recreio Dos Bandeirantes': 'Recreio dos Bandeirantes',
})
```

Abaixo verificamos quantos registros existem na base com o bairro denominado <b>Rio de Janeiro - RJ</b>

```python
df[df['bairro'] == 'Rio de Janeiro - RJ'].shape[0]
```

Considerando que bairro  uma varivel importante para a anlise e que existem apenas 28 registros com essa classificao, podemos remov-los da base de dados, o que ser feito logo abaixo:

```python
df.drop(df[df['bairro'] == 'Rio de Janeiro - RJ'].index, inplace=True)
```
