# **Analise exploratória bivariada**

A análise exploratória bivariada visa examinar a relação entre duas variáveis de um conjunto de dados. Ela se aprofunda no universo da análise univariada, onde cada variável é analisada individualmente, buscando identificar padrões, tendências e características relevantes. 

Ao analisar a relação entre as variáveis, a análise bivariada pode revelar padrões e tendências que não seriam evidentes na análise univariada. Isso pode levar à descoberta de insights importantes sobre o comportamento das variáveis e seus impactos mútuos.

A análise bivariada pode auxiliar na identificação de outliers e inconsistências nos dados, que podem comprometer a confiabilidade da análise. Essa etapa é crucial para garantir a qualidade dos dados e a robustez dos resultados.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from classes.PandasExtension import PandasExtension
from classes.utils import write_object, read_object
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Carregando a base de dados
df = pd.read_feather('./dataset/aluguel_rio_analise_bivariada.feather')
pe = PandasExtension(df)
```

```python
# função base utilizada para criar os gráfico
def chart(x, y, dat=df, sort=None, rotation=90, w=13, h=5, f=sns.barplot):
    fig, ax = plt.subplots(1, 1, figsize=(w, h))
    f(data=(dat if sort == None else dat.sort_values(sort)), x=x, y=y, ax=ax, color='#9DC098', err_kws={'color': '#316E7E'})
    sns.despine(fig=fig, ax=ax, top=True, right=True, left=False, bottom=False)
    ax.tick_params(axis='x', rotation=rotation)

# função utilizada para criar os gráfico de barra
def barplot(x, y, dat=df, sort=None, rotation=90, w=13, h=5):
    chart(x, y, dat, sort, rotation, w, h)

# função utilizada para criar os gráfico de pontos
def pointplot(x, y, dat=df, sort=None, rotation=90, w=13, h=5):
    chart(x, y, dat, sort, rotation, w, h, f=sns.pointplot)

# função utilizada para criar os gráfico de regressão
def regplot(x, y, dat=df, w=13, h=5, lowess=False, logx=False):
    fig, ax = plt.subplots(1, 1, figsize=(w, h))
    sns.regplot(data=dat, x=x, y=y, ax=ax, color='#9DC098', seed=1, lowess=lowess, logx=logx)
    sns.despine(fig=fig, ax=ax, top=True, right=True, left=False, bottom=False)
```

A escala logarítmica permite visualizar melhor as diferenças nos valores de aluguel entre os diferentes tipos de imóveis. Por exemplo, na escala logarítmica, a diferença entre o aluguel médio de um apartamento e o aluguel médio de um prédio inteiro é mais aparente do que na escala linear.

**Categorizando os imóveis por metro quadrado**

Uma abordagem prática para analisar variáveis contínuas consiste em categorizá-las e representá-las por meio de variáveis dummy. Essa estratégia simplifica a interpretação dos resultados, uma vez que cada categoria passa a ser tratada como uma categoria distinta na análise.

Agrupar os imóveis por faixas de metragem facilita a visualização da distribuição dos imóveis no mercado, por exemplo, identificando a predominância de apartamentos pequenos, médios ou grandes em diferentes regiões ou tipos de imóveis. Categorizar o metro quadrado possibilita analisar a relação entre metragem e preço dos imóveis de forma mais granular. Abaixo será feita categorização dos imóveis por metro quadrado com base no tipo de imóvel.

```python
# cria uma nova coluna do dataframe para receber a categoria referente ao metro quadrado
df['metro_quadrado_cat'] = pd.qcut(df['metro_quadrado_log'], 5)

# converte a categoria para texto para que possa ser usada em algum gráfico
df['metro_quadrado_cat'] = df['metro_quadrado_cat'].astype('str')
```
