# **Analise exploratria bivariada**

A anlise exploratria bivariada visa examinar a relao entre duas variveis de um conjunto de dados. Ela se aprofunda no universo da anlise univariada, onde cada varivel  analisada individualmente, buscando identificar padres, tendncias e caractersticas relevantes. 

Ao analisar a relao entre as variveis, a anlise bivariada pode revelar padres e tendncias que no seriam evidentes na anlise univariada. Isso pode levar  descoberta de insights importantes sobre o comportamento das variveis e seus impactos mtuos.

A anlise bivariada pode auxiliar na identificao de outliers e inconsistncias nos dados, que podem comprometer a confiabilidade da anlise. Essa etapa  crucial para garantir a qualidade dos dados e a robustez dos resultados.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from classes.PandasExtension import PandasExtension
from classes.utils import write_object, read_object
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Carregando a base de dados
df = pd.read_feather('./dataset/aluguel_rio_analise_bivariada.feather')
pe = PandasExtension(df)
```

```python
# funo base utilizada para criar os grfico
def chart(x, y, dat=df, sort=None, rotation=90, w=13, h=5, f=sns.barplot):
    fig, ax = plt.subplots(1, 1, figsize=(w, h))
    f(data=(dat if sort == None else dat.sort_values(sort)), x=x, y=y, ax=ax, color='#9DC098', err_kws={'color': '#316E7E'})
    sns.despine(fig=fig, ax=ax, top=True, right=True, left=False, bottom=False)
    ax.tick_params(axis='x', rotation=rotation)

# funo utilizada para criar os grfico de barra
def barplot(x, y, dat=df, sort=None, rotation=90, w=13, h=5):
    chart(x, y, dat, sort, rotation, w, h)

# funo utilizada para criar os grfico de pontos
def pointplot(x, y, dat=df, sort=None, rotation=90, w=13, h=5):
    chart(x, y, dat, sort, rotation, w, h, f=sns.pointplot)

# funo utilizada para criar os grfico de regresso
def regplot(x, y, dat=df, w=13, h=5, lowess=False, logx=False):
    fig, ax = plt.subplots(1, 1, figsize=(w, h))
    sns.regplot(data=dat, x=x, y=y, ax=ax, color='#9DC098', seed=1, lowess=lowess, logx=logx)
    sns.despine(fig=fig, ax=ax, top=True, right=True, left=False, bottom=False)
```

**Categorizando os imveis por metro quadrado**

Uma abordagem prtica para analisar variveis contnuas consiste em categoriz-las e represent-las por meio de variveis dummy. Essa estratgia simplifica a interpretao dos resultados, uma vez que cada categoria passa a ser tratada como uma categoria distinta na anlise.

Agrupar os imveis por faixas de metragem facilita a visualizao da distribuio dos imveis no mercado, por exemplo, identificando a predominncia de apartamentos pequenos, mdios ou grandes em diferentes regies ou tipos de imveis.
