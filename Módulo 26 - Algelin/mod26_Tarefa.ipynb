# Tarefa - Álgebra Linear

A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações

a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

b) Calcule a correlação entre os dados da **p1** e **p2**

c) Calcule a correlação entre os dados transformados (média e evolução)

d) Comente as correlações observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
```

```python
# a) Calculando média e evolução

df['média'] = df.mean(axis=1)
df['evolução'] = df.p2 - df.p1
```

```python
# b) Calcule a correlação entre os dados da p1 e p2
# c) Calcule a correlação entre os dados transformados (média e evolução)

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(1, 1, 1)

ax.axvline(0, linewidth=.5, linestyle='--')
ax.axhline(0, linewidth=.5, linestyle='--')

_ = plt.scatter(x = df.p1, y=df.p2)
_ = plt.scatter(x = df['média'], y=df['evolução'])
```

Apenas observando o gráfico podemos dizer que as notas do grupo na totalidade se mantiveram as mesmas, com metade das notas evoluindo para +2 e a outra metade para -2.

## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(1, 1, 1)

ax.axvline(0, linewidth=.5, linestyle='--')
ax.axhline(0, linewidth=.5, linestyle='--')

_ = plt.scatter(x = df.p1, y=df.p2)
_ = plt.scatter(x = (df['p1'] + df['p2']), y=df['evolução']) # eixo X somando p1 e p2
```

Substituindo a média pela soma de p1 e p2 o eixo X foi "esticado". Os pontos no gráfico referentes aos dados da matriz de transformação se mantiveram e os pontos referentes aos dados originais foram proporcionalmente "recuados" embora continuem fazendo referência aos mesmo ponto X/Y de antes.
