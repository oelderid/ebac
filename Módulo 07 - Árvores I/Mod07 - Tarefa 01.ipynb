#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
def criar_dummies(coluna: str) -> pd.DataFrame:
    dados = pd.get_dummies(data=df, columns=[coluna])
    for _, nome in enumerate(dados.columns):
        if nome.startswith(nome):
            novo_nome = nome.replace('-', '').replace('\n', ' ').replace(' / ', ' ').replace(' ', '_').lower()
            dados.rename(columns={nome : novo_nome}, inplace=True)
            
    return dados
````

````python
df['sexo'] = df['sexo'].map({'M': 0, 'F': 1})
df['posse_de_veiculo'] = df['posse_de_veiculo'].map({'N': 0, 'Y': 1})
df['posse_de_imovel'] = df['posse_de_imovel'].map({'N': 0, 'Y': 1})
````

````python
df = criar_dummies('tipo_renda')
df = criar_dummies('educacao')
df = criar_dummies('estado_civil')
df = criar_dummies('tipo_residencia')
````
