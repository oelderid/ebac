````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('previsao_de_renda.csv')
````

````python
def grafico_distribuicao(coluna: str) -> None:
    tab = pd.crosstab(
        index = df['data_ref'], 
        columns = df[coluna],
        normalize = 'index'
    )
    ax = tab.plot.bar(stacked = True, title=coluna.upper())

    tick_labs = df['data_ref'].map(lambda ts: "{}-{}".format(ts.split('-')[0], ts.split('-')[1]) ).unique()

    ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
    labels = ax.set_xticklabels(tick_labs, rotation=90)

    plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
    
def grafico_perfil_media(coluna: str) -> None:
    ax = sns.pointplot(x="data_ref", y=coluna, data=df, dodge=True, errorbar=('ci', 95))
    ax.set(title=coluna.upper())
    tick_labs = df['data_ref'].map(lambda ts: "{}-{}".format(ts.split('-')[0], ts.split('-')[1]) ).unique()
    ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
    labels = ax.set_xticklabels(tick_labs, rotation=90)
````
