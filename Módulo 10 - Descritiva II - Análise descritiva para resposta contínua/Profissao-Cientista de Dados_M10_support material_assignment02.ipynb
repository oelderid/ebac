1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('previsao_de_renda.csv')
````

````python
df['posse_de_imovel'].replace({True: 'SIM', False: 'NÃO'}, inplace=True)
df['posse_de_veiculo'].replace({True: 'SIM', False: 'NÃO'}, inplace=True)
````

````python
# posse de veículo
_ = sns.pointplot(x="posse_de_veiculo", y="renda", data=df, dodge=True, errorbar=('ci', 95), order=['SIM', 'NÃO'])
````

````python
# posse de imóvel
_ = sns.pointplot(x="posse_de_imovel", y="renda", data=df, dodge=True, errorbar=('ci', 95), order=['SIM', 'NÃO'])
````

````python
_ = sns.pointplot(
    x="posse_de_veiculo", 
    y="renda", 
    hue='posse_de_imovel', 
    data=df, 
    dodge=True, 
    errorbar=('ci', 95), 
    order=['SIM', 'NÃO']
)
````

**Conclusão:**
As duas variáveis qualitativas discriminam renda de forma estatisticamente significante pois podemos verificar no gráfico que pessoas com maior renda possuem imóvel e veículo.
